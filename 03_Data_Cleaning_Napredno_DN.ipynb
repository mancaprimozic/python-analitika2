# Naloga 1

Želimo narediti pizzo.
Obdržati želimo vse vrstice, ki imajo v stolpcu `name` vrednost pizza.

```python
# Rešitev

pizzas = recipes[recipes['name'].str.contains(pat = 'pizza', case = False, na = False)]

pizzas
```

```python
# Rešitev
sausageEggPizzas = pizzas[pizzas['ingredients'].str.contains(pat = '[Ee]gg', case = False, na = False) & pizzas['ingredients'].str.contains(pat = '[Ss]ausage', case = False, na = False)]

sausageEggPizzas
```

# Naloga 3

Želimo le pizze, ki jih je mogoče napraviti v 45min ali manj.
Želimo obdržati vrstice, ki v stolpcu `tags` vsebujejo **45-minutes-or-less** ali katerokoli manjšo številčno vrednost.

```python
pd.options.display.max_colwidth = 1000
```

```python
import re
pattern = "[0-4][0-5]-minutes-or-less"
quickPreparedPizzas = pizzas[pizzas['tags'].str.contains(pattern)]
quickPreparedPizzas
```

# Naloga 4

Sedaj nas zanima na koliko stopinj moramo razgreti pečico.
Dodati želimo stolpec **preheating** v katerega shranimo na koliko stopinj je potrebno segreti pečico. Ta podatek najdemo v stolpcu `steps`.
(Če vrednosti ni, lahko pustimo NaN).

```python
#iščemo besedo "oven" in za tem številko, ki pove, na koliko stopinj ogrejemo pečico
pattern = r'(oven?[\w+\s]{0,}[\d])'

pizzas['preheating'] = pizzas['steps'].str.extract(pattern)[0].str.split().str.get(-1)

```
